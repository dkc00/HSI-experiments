## Collection of code snippets to use numpy and pandas modules for data analysis 

```python
import random as rd 
import math 
import pandas as pd 
import numpy as np
import numpy.ma as ma
import os 
import sys
import matplotlib.pyplot as plt 
```

```python
germany = [rd.randint(5,11) for i in range(50)]
turkey = [rd.randint(1,8) for i in range (50)]
print(germany)
print(turkey)
```

## Pandas
pd.DataFrame 

pd.DataFrame(data=None, index=None, columns=None, dtype=None, copy=None)


```python
d = {'Döner in Germany [€]': germany, 'Döner in Turkey [€]': turkey} #inferred data type int64
doe = pd.DataFrame(data=d)
doe
```

```python
type(d)
# Shows me the object type of the dictionary d. d.type is the wrong code and will cause the following error:
# AttributeError: 'dict' object has no attribute 'type'
```

```python
doe.describe()
#Method used in pandas to get information about descriptive statistical parameters such as quantiles, mean, ...
```

```python
plt.scatter(doe['Döner in Germany [€]'], doe['Döner in Turkey [€]'], color='black', label='Germany')
plt.scatter(doe['Döner in Turkey [€]'], doe['Döner in Germany [€]'], color='red', label='Turkey')
plt.xlabel('Döner Kebab price in Germany [€]')
plt.ylabel('Döner Kebab price in Turkey [€]')
plt.legend()
plt.title('The Döner price comparison you all waited for')
plt.show()
```

```python
np.random.seed(123456) #Sets a certain seed which can be reproduced later in the code and always uses the same numbers

ts = pd.Series(np.random.randn(9000), index=pd.date_range("1/1/2000", periods=9000))

# np.random.randn(x) is used for random positive and negative numbers with a mean of 9 and standard deviation of 1. 
# np.randint(1,12, size = 9000) would be another idea and possible for another case.

ts = ts.cumsum() #cumulative sum 

ts.plot()
plt.title('Value change of certain Döner stocks')
plt.show()
```

```python
test = pd.Series([1, 3, 5, 7, 10])
c = test.cumsum()
c
# 1+3 = 4
# 1+3+5 = 9 
# 1+3+5+7 = 16
# 1+3+5+7+10 = 26
```
